```python
#lowering all the content
df['categoryName'] = df['categoryName'].str.lower()
df['subCategoryName'] = df['subCategoryName'].str.lower()
df['materialGroupName'] = df['materialGroupName'].str.lower()
df['brandName'] = df['brandName'].str.lower()
```

```python
columns_to_process=["categoryName","subCategoryName"]

for column in columns_to_process:
    df[column] = df[column].str.replace("&"," and ")
```

```python
#cleaning 
columns=['subCategoryName','categoryName','materialGroupName','brandName']
for col in columns:
    df[col] = df[col].str.strip()
df['subCategoryName'] = df['subCategoryName'].str.replace(r'[!@#$%^&*().,&\s]+', '-', regex=True)
df['categoryName'] = df['categoryName'].str.replace(r'[!@#$%^&*().,&\s]+', '-', regex=True)
df['materialGroupName'] = df['materialGroupName'].str.replace(r'[!@#$%^&*().,&\s]+', '-', regex=True)
df['brandName'] = df['brandName'].str.replace(r'[!@#$%^&*().,&\s]+', '-', regex=True)
```

```python
#cleaning
df['skuName'] = df['skuName'].replace(r'\s+', ' ', regex=True)
df['skuDescription'] = df['skuDescription'].replace(r'\s+', ' ', regex=True)
```
